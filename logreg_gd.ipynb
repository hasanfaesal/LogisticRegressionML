```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0,
                          n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=11)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1], c=y, cmap='winter', s=100)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression(penalty=None,solver='sag')
```

```python
logreg.fit(X,y)
```

```python
print("the values of coefficient: ", logreg.coef_)
print("the values of intercept: ", logreg.intercept_)
```

```python
m1 = -(logreg.coef_[0][0]/logreg.coef_[0][1])
b1 = -(logreg.intercept_/logreg.coef_[0][1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m1*x_input + b1
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=2)
plt.scatter(X[:,0],X[:,1],c=y, cmap='winter', s=100)
plt.ylim(-3,2)
```

```python
def gdlr(X,y):
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    learning_rate = 0.5

    for i in range(5000):
        y_hat = sigmoid(np.dot(X,weights))
        weights = weights + learning_rate*(np.dot((y-y_hat),X)/X.shape[0])

    return weights[1:],weights[0]
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
coef_,intercept_ = gdlr(X,y)
```

```python
print("the values of coefficient: ", coef_)
print("the values of intercept: ", intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input1 + b
```
